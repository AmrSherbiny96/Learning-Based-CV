```python
import os 

dataset_dir = '/content/drive/My Drive/Computer Vision/Leaves/'
train_dir = os.path.join(dataset_dir, 'training')
test_dir =  os.path.join(dataset_dir,'testing')
val_size = 0.2
```

```python
from skimage.io import imread
from skimage.transform import resize
import os
import numpy as np

def remap_labels(labels):
    '''Dataset contains labels [1,2,4,6]. Remap them to be
    in range [0-4]'''
    new_labels = np.zeros(labels.shape)
    labels_dict = {1:0, 2:1, 3:2, 4:3, 6:4}
    
    for i in range(labels.shape[0]):
        new_labels[i] = labels_dict[labels[i]] 
    
    return new_labels

def load_imgs(dir, target_shape=(128,128), shuffle=True):
    """Load images. Each directory is named after the number of leaves
    images in it conatain, so directory name can be used as the label."""
    imgs, labels = [], []
    for root, dirs, files in os.walk(dir):
        label = root.split('/')[-1]
        for f in files:
            img  = imread(os.path.join(root,f))
            img = resize(img, target_shape)
            imgs.append(img)
            labels.append(int(label))

    imgs_arr = np.stack(imgs)

    labels_arr = np.array(labels)

    if(shuffle):
        indices = np.arange(imgs_arr.shape[0])
        np.random.shuffle(indices)
        imgs_arr = imgs_arr[indices]
        labels_arr = labels_arr[indices]
    return imgs_arr, labels_arr    
```

```python
#load training and testing images. Remap labels to be in range 0-4

train_imgs, train_labels = load_imgs(train_dir)
train_labels = remap_labels(train_labels)

test_imgs, test_labels = load_imgs(test_dir)
test_labels = remap_labels(test_labels)
```

```python
#split training data to training and validation

n_imgs = train_imgs.shape[0]
val_idxs = np.random.choice(np.arange(n_imgs), size=int(val_size * n_imgs),
                            replace=False)

val_imgs, val_labels = train_imgs[val_idxs], train_labels[val_idxs]
train_imgs = train_imgs[~np.isin(np.arange(n_imgs), val_idxs)]
train_labels = train_labels[~np.isin(np.arange(n_imgs), val_idxs)]
```

```python
import matplotlib.pyplot as plt

# Visualize some samples from the data

fig, ax = plt.subplots(5, 1, figsize=(5, 10))
for i in range(5):
    img = train_imgs[train_labels == i][0]
    ax[i].imshow(img)
    ax[i].set_xticks([])
    ax[i].set_yticks([])
        
fig.tight_layout()
```

## Simple Classification Network - First 2 blocks of VGG-16
Using the 2 pretrained blocks of VGG-16 and adding more blocks to be trained, we build a simple classification network.

```python
from tensorflow import keras
from keras.models import Model
from keras import layers
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.vgg16 import VGG16, preprocess_input

vgg_model = VGG16(input_shape=(128, 128, 3), include_top=False)
cnn_model = keras.Sequential()
vgg_model.trainable = False

for layer in vgg_model.layers[:7]:
    cnn_model.add(layer)

cnn_model.add(layers.Conv2D(128, (3, 3), padding="same", activation="relu"))
cnn_model.add(layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
cnn_model.add(layers.Conv2D(256, (3, 3), padding="same", activation="relu"))
cnn_model.add(layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

cnn_model.add(layers.Flatten())
cnn_model.add(layers.Dense(128, activation='relu'))
cnn_model.add(layers.Dense(64, activation='relu'))

cnn_model.add(layers.Dense(5, activation='softmax'))

cnn_model.summary()
```

```python
from tensorflow.keras.optimizers import RMSprop
import tensorflow as tf

early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_acc', mode='max', patience=20)

checkpoint_filepath = 'best_cnn_weights.hdf5'
model_checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_filepath,
    save_best_only=True)

cnn_model.compile(loss='sparse_categorical_crossentropy',
              optimizer=RMSprop(learning_rate=1e-4),
              metrics=['acc'])

history = cnn_model.fit(
      x=train_imgs,
      y=train_labels,
      epochs=100,
      validation_data=(val_imgs, val_labels),
      verbose=1,
      callbacks=[early_stop, model_checkpoint])
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

def plot_conf_mat(conf_mat, labels=[1,2,3,4,6]):
    df = pd.DataFrame(conf_mat, index = [i for i in labels],
                  columns = [i for i in labels])
    plt.figure(figsize = (10,7))
    sns.heatmap(df, annot=True)

def visualize_results(history, acc_metric='acc'):
    acc = history.history[acc_metric]
    val_acc = history.history['val_'+acc_metric]
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs = range(len(acc))

    plt.plot(epochs, acc, 'b', label='Training acc')
    plt.plot(epochs, val_acc, 'r', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()

    plt.figure()

    plt.plot(epochs, loss, 'b', label='Training loss')
    plt.plot(epochs, val_loss, 'r', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()

    plt.show()
```

## Reformulating the Problem as a Regression Problem.

Using the same network from the previous step, we change the output layer to predict a continuous number instead of a discrete class.

```python
from tensorflow.keras.metrics import Accuracy

# Custom accuracy function to read labels
def custom_acc(y_true, y_pred):
    """Rounds prediction and then computes accuracy."""
    y_pred = tf.math.round(y_pred)
    acc = tf.keras.backend.equal(y_true,y_pred)
    mean_acc = tf.math.reduce_mean(tf.cast(acc, tf.float32))
    return mean_acc
```

```python
from tensorflow import keras
from keras.models import Model
from keras import layers
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.vgg16 import VGG16, preprocess_input

vgg_model = VGG16(input_shape=(128, 128, 3), include_top=False)
cnn_reg_model = keras.Sequential()
vgg_model.trainable = False

for layer in vgg_model.layers[:7]:
    cnn_reg_model.add(layer)

cnn_reg_model.add(layers.Conv2D(128, (3, 3), padding="same", activation="relu"))
cnn_reg_model.add(layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
cnn_reg_model.add(layers.Conv2D(256, (3, 3), padding="same", activation="relu"))
cnn_reg_model.add(layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

cnn_reg_model.add(layers.Flatten())
cnn_reg_model.add(layers.Dense(64, activation='relu'))

cnn_reg_model.add(layers.Dense(1, activation='linear'))

cnn_reg_model.summary()
```

```python
from tensorflow.keras.optimizers import RMSprop
import tensorflow as tf

early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_custom_acc', mode='max', patience=20)
checkpoint_filepath = 'best_reg_weights.hdf5'

model_checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_filepath,
    save_best_only=True)

cnn_reg_model.compile(loss='mean_squared_error', optimizer= RMSprop(learning_rate=0.0001),
                      metrics=[custom_acc])
      
history = cnn_reg_model.fit(
      x=train_imgs,
      y=train_labels,
      epochs=100,
      validation_data=(val_imgs, val_labels),
      callbacks=[early_stop, model_checkpoint])
```

```python
def round_preds(preds, values=[0,1,2,3,4]):
    rounded_preds = np.zeros(preds.shape)
    for i in range(preds.shape[0]):
        rounded_preds[i] = (np.abs(preds[i] - values)).argmin()
    return rounded_preds
```

```python
best_model = tf.keras.models.load_model('best_reg_weights.hdf5', custom_objects={'custom_acc':custom_acc})

train_preds = best_model.predict(train_imgs)
rounded_train_preds = round_preds(train_preds)

val_preds = best_model.predict(val_imgs) 
rounded_val_preds = round_preds(val_preds)

test_preds = best_model.predict(test_imgs)
rounded_test_preds = round_preds(test_preds)
```

# Discussion

To train the network, I mapped the labels to be in range 0-4 instead of 1-6 to allow using sparse categorical crossentropy.

In the first part of classification, the training loss decreases and the accuracy almost reaches 100%, whereas the validation accuracy heavily oscillates around a maximum of 50%. This can be due to several reasons. First of all, we used only the first two blocks of VGG16, which may learn general features that aren't enough to discriminate between classes. After that, we add a few convolution and dense layers, but the number of parameters is very large and the data is small, not allowing the network to capture the actual difference in the classes and causing overfitting.

In the second part where we train a regression model instead to predict our classes. To do so, I tried different approaches. In the first time, I implemented a variant of ReLU that flattens at the maximum label of 4. This was obviously a bad idea since the function has 0 gradiant almost everywhere, causing it to get stuck very easily. To account for this, I used a linear activation function insead, and let the network learn the boundaries on its own. At prediction, I rounded the predictions to the closes available label. The training accuracy was again very high while validation and testing accuracies where around 38%, as the network is not employing any regularization and the dataset is small, causing overfitting.

## Adding regularization

```python
from tensorflow.keras import activations

vgg_model = VGG16(input_shape=(128, 128, 3), include_top=False)
cnn_regularized_model = keras.Sequential()
vgg_model.trainable = False

for layer in vgg_model.layers[:7]:
    cnn_regularized_model.add(layer)

cnn_regularized_model.add(layers.Conv2D(128, (3, 3), padding="same"))
cnn_regularized_model.add(layers.BatchNormalization())
cnn_regularized_model.add(layers.Activation(activations.relu))
cnn_regularized_model.add(layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

cnn_regularized_model.add(layers.Conv2D(256, (3, 3), padding="same"))
cnn_regularized_model.add(layers.BatchNormalization())
cnn_regularized_model.add(layers.Activation(activations.relu))
cnn_regularized_model.add(layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

cnn_regularized_model.add(layers.Flatten())
cnn_regularized_model.add(layers.Dense(128, activation='relu'))
cnn_regularized_model.add(layers.Dropout(0.25))
cnn_regularized_model.add(layers.Dense(64, activation='relu'))
cnn_regularized_model.add(layers.Dropout(0.25))
cnn_regularized_model.add(layers.Dense(5, activation='softmax'))

cnn_regularized_model.summary()
```

```python
tf.keras.backend.clear_session()

early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_acc', mode='max', patience=20)

checkpoint_filepath = 'best_regularized_weights.hdf5'
model_checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_filepath,
    save_best_only=True)

cnn_regularized_model.compile(loss='sparse_categorical_crossentropy',
              optimizer=keras.optimizers.RMSprop(learning_rate=1e-4),
              metrics=['acc'])

history = cnn_regularized_model.fit(
      x=train_imgs,
      y=train_labels,
      epochs=100,
      validation_data=(val_imgs, val_labels), 
      verbose=1,
      callbacks=[early_stop, model_checkpoint])
```

### Using Data Augmentation

```python
from tensorflow.keras import activations

vgg_model = VGG16(input_shape=(128, 128, 3), include_top=False)
vgg_model.trainable = False

cnn_aug_model = tf.keras.Sequential([layers.InputLayer(input_shape=(128,128,3)),
                                     layers.RandomFlip("horizontal_and_vertical"),
                                      layers.RandomRotation(0.4),
                                      #layers.RandomHeight(factor=0.3)
                                      layers.RandomTranslation(height_factor=0.3, width_factor=0.3)])


for layer in vgg_model.layers[:7]:
    cnn_aug_model.add(layer)

cnn_aug_model.add(layers.Conv2D(128, (3, 3), padding="same"))
cnn_aug_model.add(layers.BatchNormalization())
cnn_aug_model.add(layers.Activation(activations.relu))
cnn_aug_model.add(layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

cnn_aug_model.add(layers.Conv2D(256, (3, 3), padding="same"))
cnn_aug_model.add(layers.BatchNormalization())
cnn_aug_model.add(layers.Activation(activations.relu))
cnn_aug_model.add(layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

cnn_aug_model.add(layers.Flatten())
cnn_aug_model.add(layers.Dense(128, activation='relu'))
cnn_aug_model.add(layers.Dropout(0.25))
cnn_aug_model.add(layers.Dense(64, activation='relu'))
cnn_aug_model.add(layers.Dropout(0.25))
cnn_aug_model.add(layers.Dense(5, activation='softmax'))

cnn_aug_model.summary()
```

```python
tf.keras.backend.clear_session()

early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_acc', mode='max', patience=20)

checkpoint_filepath = 'best_aug_weights.hdf5'
model_checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_filepath,
    save_best_only=True)

cnn_aug_model.compile(loss='sparse_categorical_crossentropy',
              optimizer=keras.optimizers.RMSprop(learning_rate=1e-4),
              metrics=['acc'])

history = cnn_aug_model.fit(
      x=train_imgs,
      y=train_labels,
      epochs=100,
      validation_data=(val_imgs, val_labels), 
      verbose=1,
      callbacks=[early_stop, model_checkpoint])
```

# Discussion

In the second part, using regularization of dropout and batch normalization helped speed up the training, and gave a boost of 7% on the validation accuracy to reach 57%. The learning curve was still not smooth as validation accuracy heavily fluctuates, and the training accuracy again reached almost 100%. This is due to the very small data size with respect to the large number of parameters.

Finally, I used data augmentation to help the network better generalize and increase the available data. To do so, I used keras layers to perform random flips, random rotation, and random translations within predefined ranges. This reduced the training accuracy heavily to reach a maximum of 50% meaning that the network is no longer overfitting the training data, but didn't help boost the validation and testing accuracy, as it stayed at around 57% as in the previous step. Perhaps different augmentation techniques like blurring or adjusting brightness may be of more help.
